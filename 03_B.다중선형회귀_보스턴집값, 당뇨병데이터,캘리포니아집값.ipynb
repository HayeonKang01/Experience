# 보스턴 집값 다중선형회귀 분석(사이킷런)

```python
# 데이터 준비, 탐색은 위에서 수행한 것으로 대체
```

### 데이터 준비

#### 데이터 분할

```python
from sklearn.model_selection import train_test_split

# X(독립변수), Y(종속변수) 분할 
# 아까 찰스때는 inplace=True를 줬었는데, 이번엔 안넣었다 왜?
# -> 인플레이스 투르 안썼기때문에 원래 데이터가 유지된다. inplace=True를 하면 그 드롭한 결과가 본 데이터에서 저장된다.
X = boston_df.drop(['PRICE'], axis = 1)
y = boston_df['PRICE']

# 학습용 데이터와 평가용 데이터로 분할하기 
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#X_train 데이터 일부 확인
X_train.head()
```

### 모델 클래스 선택

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

### 모델 구축 

```python
#학습 수행
reg.fit(X_train, y_train)
```

### 모델 평가

```python
from sklearn.metrics import mean_squared_error, r2_score

# 평가(검증) 데이터로 예측 수행 -> 예측 결과 y_pred구하기
y_pred = reg.predict(X_test)

# 평가 지표 계산
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print("MSE: {0:.3f}, RMSE: {1:.3f}".format(mse,rmse))
print("R2: {:.3f}".format(r2))
```

```python
# 수정된 결정 계수 확인(Ajusted-R2)

#데이터의 크기
n = len(X_train)

#독립변수의 갯수
k = len(X_train.columns)

# 수정된 결정 계수
adj_r2 = 1 - ((1-r2) * (n-1) / (n-k-1))

print("Adjusted R2: {:.4f}".format(adj_r2))
```

##### Adj R2가 R2 보다 약간 작다

#### 회귀 계수 확인 : 각각의 독립 변수의 설명력 확인

```python
#회귀  모델에서 구한 회귀 계수와 피처 이름을 묶어 
#시리즈 자료형으로 만듦
```

```python
# 회귀계수확인: 각각의 독립변수의 설명력 확인

# ndarray의 실수를 부동 소수점으로 출력 

#회귀계수와 절편값 확인 
print("Y 절편: {:.3f}".format(reg.intercept_))
print("회귀계수: ", np.round(reg.coef_,3))

# ndarray의 실수 출력 옵션 복구 
np.set_printoptions(suppress=False)
```

```python
# 회귀모델에서 구한 회귀계수와 피처이름을 묶어 시리즈 자료형으로 만듦 
coef = pd.Series(data = np.round(reg.coef_, 3),\
                index = X.columns)

print('y = '+str(reg.intercept_)+' +')

for i, c in zip(coef.index, coef.values): # 시리즈에 인덱스와 값을 같이 가져왔기 때문에 zip을 해서 쌍으로 하나씩 가져온다.
    print('('+str(c)+')*' + str(i))
```

##### 회귀식을 출력한 결과이다.

###  실제값과 예측값의 분포차이를 시각화

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, s=80) # x 축에 실제값, y 축에 예측값
plt.plot([5, 50], [5, 50], c='r', ls='--')
plt.xlabel("Actual price ($1,000s)")
plt.ylabel("Predcited price")
plt.grid()
plt.show()
```

#### 빨간색 선 위에 점이 찍히면 그 점들은 실제값과 예측값이 일치하는 것들이다. 
#### 빨간색 선 주변에 있는 애들은 실제값과 예측값이 굉장히 가까운 애들이다.
#### 빨간색 주변에 많이 모일수록 잘 된 모델이다. 

# 보스턴 집값: 다중선형회귀 분석(스탯츠모델)

```python
# 최소 제곱법으로 객체 생성, 학습, 예측, 평가를 진행한다.
import statsmodels.api as sm 

# 훈련 데이터로 학습 수행
X_train = sm.add_constant(X_train) # X train, y_train에 모두 add_constant 해야함
reg = sm.OLS(y_train, X_train).fit() # 객체를 생성과 학습을 같이했다. // 학습완료 
# 학습완료하면서 coef, intercept 모두 찾았다.

# 검증 데이터로 예측 수행
X_test = sm.add_constant(X_test)
y_pred = reg.predict(X_test)

#평가 지표 값 확인
print(reg.summary()) # 스탯츠에만 summary 있음
```

##### Dep.Variable : 종속변수 (PRICE) 
#### 회귀모델은 OLS 썼고 
#### method는 least squared 최소제곱법을 썼어 
#### R-squared, adj-R : adj가 약간작다.
##### F 검정 : 모델 전체의 유의성 (모델의 유의성 )
##### t 검정 : 계수의 유의성 (아래 표의 t값들)
##### p value가  0.05보다 크면 신뢰할 수 없다. 
##### coef 상수 
##### 아래 표 
- 0.05보다 큰 것 : AGE, INDUS -> 이것들의 회귀계수는 믿을만 하지 않다. -> 독립변수 준비부터 다시돌아감 
    --> 모델을 튜닝하는 것이다. (되돌아가면서)

### 데이터 준비

```python
from sklearn.model_selection import train_test_split 
```

```python
# X(독립변수), Y(종속변수) 분할
X = diab.data # 애초에 2차원
y = diab.target # 애초에 1차원 
# --> reshape할 필요 없다. (** cf. 독립은 2차원, 종속은 1차원이어야 한다.)

# 학습용 데이터와 평가용 데이터로 분할하기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

```

# ====================================

```python
### 모델 구축
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# 회귀 모델 객체 생성
reg = LinearRegression()


#학습 수행
reg.fit(X_train, y_train)

# 회귀계수 및 회귀식 
coef = pd.Series(data = np.round(reg.coef_, 3), index = diab.feature_names)
print(coef)

    
# 평가(검증) 데이터로 예측 수행 -> 예측 결과 y_pred구하기
y_pred = reg.predict(X_test)

### 모델 평가
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

n = len(X_train)
k = len(diab.feature_names) # **
adj_r2 = 1 - ((1-r2) * (n-1) / (n-k-1))

print(mse, rmse, r2, adj_r2)


```

```python
plt.figure(figsize = (5,5))
plt.scatter(y_test, y_pred, marker = '^', s=12)
plt.xlabel("Actual value")
plt.ylabel("Predcited Value")
plt.grid()
plt.show()
```

### 예측치와 실제값이 같은 것들이 잘맞춘 것 
##### 결정계수가 0.5 정도였는데 이정도 퍼져있으면 0.5구나! 라고 알면 됨
