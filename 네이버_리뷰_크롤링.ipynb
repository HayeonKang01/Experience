```python
from selenium import webdriver
from selenium.webdriver import ActionChains  # 액션체인 활성화
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

def naver_crw(link, review_page, name):

    # webdriver_manager ver.
    dr = webdriver.Chrome(ChromeDriverManager().install())

    #dr = webdriver.Chrome('/Users/hayeon/Downloads/chromedriver_mac_arm64/chromedriver')
    dr.get('{0}'.format(link))


    #driver = webdriver.Chrome()

    #res = requests.get('https://search.shopping.naver.com/catalog/32773809619?query=%EC%A0%84%ED%86%B5%EC%A3%BC&NaPm=ct%3Dlf96l4jk%7Cci%3D5e2014caf351f583be723ebf02a3fea6175d592f%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Dc0983895f1af21459ed548df8b7e361fc1551d82')
    #soup = BeautifulSoup(res.content, 'html.parser')

    act = ActionChains(dr)  # 드라이버에 동작을 실행시키는 명령어를 act로 지정

    date = []
    review = []
    title =[]

    start = time.time()
    ### 예시 ###
    
    if review_page >= 2000: 
        review_page = 101
    
    else:
        review_page = (review_page%20)+1
    
    print('REVIEW_PAGE : ', review_page)
    
    for i in range(1, review_page): # 리뷰 최대 2000개 까지라고 해서 100p로 제한
        print(i)

        #dr.execute_script("window.scrollTo(0, 10)")
        #time.sleep(10)

        if i == 1 : 
            pass

        elif i % 10 == 1: # 11, 21, 31... 
            #element2 = 
            dr.find_element_by_css_selector('#section_review > div.pagination_pagination__JW7zT > a.pagination_next__3_3ip').send_keys(Keys.ENTER)  # 동작 할 요소 선택
            #act.click(element2).perform()
            time.sleep(3)

        elif i > 11 :
            #element2 = dr.find_element_by_css_selector('#section_review > div.pagination_pagination__JW7zT > a.pagination_next__3_3ip')  # 동작 할 요소 선택
            #act.click(element2).perform()
            #time.sleep(3) 

            if (i>11) and (i%10 != 1):
                if i%10 != 0: 
                    num = i%10 +1
                elif i%10 == 0:
                    num = 11

                element1 = dr.find_element_by_css_selector('#section_review > div.pagination_pagination__JW7zT > a:nth-child({0})'.format(num)).send_keys(Keys.ENTER)  # 동작 할 요소 선택
                #act.click(element1).perform()
                time.sleep(3)

        else:
            #element1 = 
            dr.find_element_by_css_selector('#section_review > div.pagination_pagination__JW7zT > a:nth-child({0})'.format(i)).send_keys(Keys.ENTER)  # 동작 할 요소 선택
            #act.click(element1).perform()  # element1  클릭 동작을 수행
            time.sleep(3)



        mydata_date = dr.find_elements_by_css_selector('div.reviewItems_etc_area__3VUjt')
        mydata_review = dr.find_elements_by_css_selector('p.reviewItems_text__XrSSf')
        mydata_title = dr.find_elements_by_css_selector('em.reviewItems_title__AwHcz')

        for d, r, t in zip(mydata_date, mydata_review, mydata_title):
            #print(d.text, r.text, t.text, '\n')
            date.append(d.text)
            review.append(r.text)
            title.append(t.text)

        time.sleep(3)


    print(date, '\n', title, '\n', review)
    df = pd.DataFrame(np.column_stack([list(date), list(title), list(review)]), columns=['date', 'title', 'review'])
    df.to_csv('{0}.csv'.format(name))
    end = time.time()

    print(f"{end - start:.5f} sec=========================")
```
