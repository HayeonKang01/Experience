```python
tr = [['우유','계란','커피'],
     ['빵','우유','버터','휴지','라면'],
     ['맥주','땅콩'],
     ['계란','빵','우유','생수'],
     ['햇반','커피','간장','계란']]
```

```python
import apyori as ap

result = ap.apriori(tr, min_support=0.3, \
                   min_confidence = 0.6)
rules = list(result) # 결과의 반환형식을 리스트로 꼭 바꿔주어야 한다. ***
```

```python
for r in rules: 
    print(r, "\n") # 중간 공백 만들어줌

# 큰 세 덩어리로 이루어져 있다. 
```

##### 여러개의 연관규칙 중 세번째만 본다 (인덱스 번호가 2니까 세번째)
##### 빈발항목집합의 길이가 1인건 제외해야됨. (연관규칙 만들 수 없으니까)

```python
r = list(rules[2]) # rules의 2번째 {계란, 우유}를 r에 넣어준다.

print("빈발항목집합 : ", r[0]) # 0번 인덱스는 항목집합을 보여주는 부분
print("지지도: ", r[1]) # 1번 인덱스는 지지도
print("연관규칙 정보 : ", r[2]) # 연관규칙이 뭔지, 신뢰도, 향상도가 모여있는 2
print("연관규칙의 개수 :", len(r[2]))
```

```python
s = list(rules[2][2]) # 세번째의 연관규칙정보 
# 한개의 항목집합에는 여러개의 연관규칙 있을 수 있따. 때문에 s에는 여러 연관규칙 있을 수 있음

for i in s:
    print(i, "\n")
```

#### 아이템 베이스 뒤에는 선행절 
#### 아이템 애드 뒤에는 후행절

#### 이 연관규칙의 신뢰도는 0.667
#### 이 연관규칙의 향상도는 1.1111 
### 계란, 우유 라는 빈발항목집합(아이템즈)에서 나온 결과이다

#### 우유 -> 계란 연관규칙 
#### 지금 보고 있는 데이터를 꾸며서 볼 수 있으면 좋을 것 같다.

```python
p = list(rules[2][2][0])

print("선행절: ", p[0])
print("후행절 : ", p[1])
print("신뢰도: ", p[2])
print("향상도: ", p[3])
```

```python
for r in rules:
    if len(r[0]) > 1: # 항목집합의 길이가 1보다 큰것만 골라라 
        support = r[1]
        for i in list(r[2]): # 연관규칙 덩어리가 r[2]
            print("연관규칙 :", set(i[0]), "==>", set(i[1])) # 선행, 후행 가져와서 집합으로 만들어라 (대괄호 생기게)
            print("지지도 : ",round(support, 4))
            print("신뢰도 : ", round(i[2], 4))
            print("향상도 : ", round(i[3], 4))
            print()
```

## 샘플을 간편하게 50개 뽑기

```python
# 트랜잭션의 개수 총 수 
tr_num = 50

# 한 트랜잭션 내의 항목의 개수  (범위)
tr_len = list(range(2, 6))

# 항목이 가질 수 있는 값의 범위 
item_domain = list(range(1,21)) # 상품아이디가 1부터 20까지  
```

```python
item_domain
```

```python
data = [] # 얘 리스트, 트랜잭션도 리스트니까 트랜잭션 넣으면 이차원 됨

for a in range(tr_num):
    c = random.sample(item_domain, random.choice(tr_len)) # 범위, 개수
    c.sort()
    data.append(c)
```

```python
data
```

# 연관규칙 탐사

```python
import apyori as ap

rules = list(ap.apriori(data, min_support = 0.1, min_confidence = 0.2)) # 최소신뢰도는 보통 최소지지도보다 좀 높게 한다.
```

#### 리스트 변환 꼭 해야한다. 

```python
rules # 너무 안보이니까 잘 보이게 바꾸자
```

```python
for r in rules:
    if len(r[0]) > 1:
        support = r[1]
        ar = list(r[2])
        for a in ar:
            print(set(a[0]), " ==>", set(a[1]))
            print("support = ", r[1])
            print('confidence = ', a[2])
            print("lift = ", a[3])
            print()
```

```python
import pandas as pd 

result = []

for r in rules:
    if len(r[0]) > 1 :
        support = r[1]
        ar = list(r[2]) #연관규칙 덩어리
        for a in ar:
            row = [set(a[0]), "==>", set(a[1]), support, round(a[2],4), round(a[3], 4)] # 한 행이다. 
            result.append(row)

col = ["선행절", "==>", "후행절", "지지도","신뢰도","향상도"]
df = pd.DataFrame(result, columns=col, index = range(1, len(result)+1))
```
